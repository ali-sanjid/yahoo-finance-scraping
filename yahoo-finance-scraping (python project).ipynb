```python
import requests
import pandas as pd
from io import StringIO
import csv
from tqdm import tqdm_notebook as tqdm
```

```python
company_list = ['AMZN', 'TSLA', 'AAPL'] # this list containts desired stock data names.you can add or delete any stock names from the list
stockdata = [] #desired stock data will be appended to this list
def getData(company_stock):
    url = f'https://query1.finance.yahoo.com/v7/finance/download/{company_stock}?'
    #created parameters to find specific data range
    params = {                        
           'range': '1y',
           'interval':'1mo',                          
           'events':'history',
           'includeAdjustedClose':'true'}           
    url_info = requests.get(url,params = params, headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"})
    #importing all the information and converting it
    file = StringIO(url_info.text)
    reader = csv.reader(file)
    data = list(reader)
    
    #creating dataFrame with custom column headers
    data_frame = pd.DataFrame(data[1:], columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
    data_frame.insert(1, 'Company Name', f'{company_stock}')
    return pd.DataFrame(data_frame)

#Appending the stock data individually
    
for company in company_list:
    stockdata.append(getData(company))
    print(f'Getting: {company}')

#concatenating all the individual data
combined_list = pd.concat(stockdata)
print(combined_list)
```

```python
#from the combined_list, it can be cleaned easily according to the required information. Suppose we only want the Date, company and closing price. We can use the below:
closing_price = combined_list[['Date', 'Company Name', 'Close']]
print(closing_price)
```
